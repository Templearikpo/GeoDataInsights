# Group cars by manufacturers and analyze their average price and fuel efficiency.

```python

# Group by Manufacturer and calculate the average price and fuel efficiency
average_price_fuel_efficiency = df.groupby('Manufacturer').agg(
    average_price=('Price_in_thousands', 'mean'),
    average_fuel_efficiency=('Fuel_efficiency', 'mean')
).reset_index()

print(average_price_fuel_efficiency)
```

```python
# Sort by average price or fuel efficiency 
sorted_by_price = average_price_fuel_efficiency.sort_values(by='average_price', ascending=False)
print(sorted_by_price)

plt.figure(figsize=(12, 6))
```

```python
# Bar plot for average price
plt.figure(figsize=(24, 12))
plt.subplot(1, 2, 1)
plt.bar(sorted_by_price['Manufacturer'], sorted_by_price['average_price'], color='skyblue')
plt.title('Average Price by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Price (in thousands)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot for average fuel efficiency
plt.figure(figsize=(24, 12))
plt.subplot(1, 2, 2)
plt.bar(sorted_by_price['Manufacturer'], sorted_by_price['average_fuel_efficiency'], color='salmon')
plt.title('Average Fuel Efficiency by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Fuel Efficiency (mpg)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Which car model has the highest resale value compared to its initial price?

df['resale_to_price_ratio'] = df['year_resale_value'] / df['Price_in_thousands']

# Sort the dataset based on this ratio in descending order
sorted_by_ratio = df.sort_values(by='resale_to_price_ratio', ascending=False)

# the car model with the highest resale-to-price ratio
highest_resale_model = sorted_by_ratio[['Model', 'Manufacturer', 'year_resale_value', 'Price_in_thousands', 'resale_to_price_ratio']].iloc[0]

print(f"The car model with the highest resale value compared to its initial price is:")
print(highest_resale_model)
```

```python
#Find the top 3 fuel-efficient cars with an engine size above 2.5 liters.

# Filter cars with engine size greater than 2.5 liters
filtered_cars = df[df['Engine_size'] > 2.5]

# Sort by fuel efficiency in descending order (more efficient cars at the top)
sorted_by_fuel_efficiency = filtered_cars.sort_values(by='Fuel_efficiency', ascending=False)

#Select the top 3 most fuel-efficient cars
top_3_fuel_efficient_cars = sorted_by_fuel_efficiency[['Model', 'Manufacturer', 'Engine_size', 'Fuel_efficiency']].head(3)

print("Top 3 fuel-efficient cars with an engine size above 2.5 liters:")
print(top_3_fuel_efficient_cars)
```

```python
# Filter the dataset for Lexus models
lexus_models = df[df['Manufacturer'] == 'Lexus']

# Sort the Lexus models by horsepower in descending order
sorted_lexus_models = lexus_models.sort_values(by='Horsepower', ascending=False)

# Select the Lexus model with the highest horsepower
lexus_highest_horsepower = sorted_lexus_models[['Model', 'Horsepower']].head(1)

print("Lexus model with the highest horsepower:")
print(lexus_highest_horsepower)
```
